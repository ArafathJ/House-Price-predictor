```python
import pandas as pd
import numpy as np
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=6)
```

```python
X_train = pd.read_csv('../Data/X_train.csv')
y_train = pd.read_csv('../Data/y_train.csv')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, Y_train, Y_test = train_test_split(X_train, y_train, test_size=0.1, random_state = 4)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model.fit(x_train, Y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
y_hat = model.predict(x_test)
```

```python
score = r2_score(Y_test, y_hat)
```

```python
print(score)
```

```python
score_ = mean_squared_error(y_hat, Y_test)
```

```python
score_
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=100, max_depth=10)
```

```python
rfr.fit(x_train, Y_train)
```

```python
y_ = rfr.predict(x_test)
```

```python
print(r2_score(Y_test, y_))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gradient_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=6)
```

```python
gradient_model.fit(x_train,Y_train)
```

```python
y_ = gradient_model.predict(x_test)
```

```python
print(r2_score(y_, Y_test))
```
