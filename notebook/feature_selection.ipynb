```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#for feature selection we uses 2 libraries
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../Data/preprocessed_train_data.csv')
```

```python
df.head()
```

```python
y_train = df[['SalePrice']]
X_train = df.drop(['SalePrice', 'Id'], axis=1)
```

```python
## apply the feature selection

# first , I specify the Lasso Regression models and I
# select a suitable alpha (equivalent of penalty)
# The bigger the alpha the less features that wil be selected 

#Then I use SelectFromModel object from sklearn 
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_support()
```

```python
### geting info from the feature_sel_model

selected_feature = X_train.columns[(feature_sel_model.get_support())]

print(f"Selected features are : {selected_feature}")
print(f"count of them are :{len(selected_feature)}")
```

```python
X_train.shape[1]
```

```python
X_train = X_train[selected_feature]
```

```python
X_train.head()
```

```python
X_train.to_csv('../Data/X_train.csv',index = False)
```
